## Part 6: Reflection (10 min)

Answer each question in **3-5 sentences**. Connect your answers to concepts from Chapter 25.

### Q1: What worked well with AI-assisted development?

*Think about: speed of scaffolding, code quality, design suggestions, handling of repetitive tasks.*

**Your answer:** AI-assisted development worked, Claude Code quickly scaffolded a complete portfolio structure with sections like Hero, Projects, Skills, Contact, and Footer. Instead of writing every file from scratch, I could describe the professional data science and economics style I wanted, and the AI generated a working Next.js layout with reusable components. It was especially helpful for repetitive tasks such as creating project cards, skill tags, buttons, and consistent styling. It helped speed up the “Action” stage of development, letting me test design ideas faster.



### Q2: What did you have to fix manually?

*Think about: layout bugs, incorrect assumptions, styling mismatches, missing features, hallucinated APIs.*

**Your answer:** I had to fix several things manually because the AI made some assumptions that did not fully match my background or local setup. For example, it included “JavaScript / TypeScript” in the Skills section, but I wanted the portfolio to focus more on economics, data science and other parts like Python, R, SQL, and causal inference. I also had to deal with local development issues, including WSL permissions, npm installation problems, and GitHub authentication. These problems showed that AI can generate code quickly, but the user still needs to check whether the result is technically correct and personally accurate.



### Q3: How does AI-assisted development compare to writing code entirely yourself?

*Think about: the HADI loop from Chapter 25 (Hypothesis, Action, Data, Insight). Did you follow this pattern? Where did you need domain knowledge that the AI lacked?*

**Your answer:** Compared to writing code entirely myself, AI-assisted development felt much faster, but it also required more evaluation and correction. I formed a hypothesis about what I wanted, asked Claude Code to take an action, checked the output in the browser, and then used the result as data to decide what to change next. For example, after seeing the first Skills section, I realized some skills did not signal my intended identity as an economics and data science student. My own domain knowledge was necessary because the AI did not automatically know which skills, projects, and design choices best represented my academic and career goals.



### Q4: Prompt engineering lessons

*What makes a good prompt for Claude Code? What makes a bad one? Give one specific example of a prompt you improved.*

**Your answer:** A good prompt for Claude Code is specific about the goal, the design style, the content, and what should not be included. A bad prompt is too vague, such as “make it better,” because the AI may make random design choices or add features that do not match the portfolio purpose. One prompt I improved was the design customization prompt: instead of only saying “make it professional,” I specified deep navy blue, soft blue, warm off-white, dark slate text, white cards, soft shadows, rounded corners, and a clean academic style. That made the output much closer to the portfolio identity I wanted.



### Q5: If you were hiring a junior data scientist, would their portfolio site influence your evaluation? Why or why not?

*Connect to the "signaling" concept from labor economics — what does a portfolio signal that a resume alone does not?*

**Your answer:** Yes, a portfolio site would influence my evaluation because it gives evidence beyond a resume. As the concept of signaling identifies, a portfolio signals skills, effort, taste, and the ability to complete real projects in a visible way. A resume can list the different skills one has but a portfolio can show actual project descriptions, tools used, methods applied, and communication ability. For a junior data scientist, I would not judge only by the website design, but a clear and well-organized portfolio would make the candidate look more credible and prepared.
